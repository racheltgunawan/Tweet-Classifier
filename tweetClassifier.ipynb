```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/archive-2/train.csv')

#remove punctuation
def remove_punctuation(text):
    final = "".join(u for u in text if u not in ("?", ".", ";", ":",  "!",'"'))
    return final
df['tweet'] = df['tweet'].apply(remove_punctuation) 

#split datafile
index = df.index
import numpy as np
df['random_number'] = np.random.randn(len(index))
train = df[df['random_number'] <= 0.8]
test = df[df['random_number'] > 0.8]

train.head()
```

```python
test.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

#download stopwords from nltk 
import nltk
#from nltk.corpus import stopwords
nltk.download('stopwords')

#also gets rid of stopwords
vectorizer = CountVectorizer(stop_words = stopwords.words('english'), token_pattern=r'\b\w+\b')

#fit train
train_matrix = vectorizer.fit_transform(train['tweet'])
#print(vectorizer.get_feature_names())
test_matrix = vectorizer.transform(test['tweet'])
```

```python
#use logistic regression model
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
#split datasets and labels for train and test
x_train = train_matrix
x_test = test_matrix
y_train = train['label']
y_test = test['label']
```

```python
#fit trains
lr.fit(x_train, y_train)
```

```python
#predict with model
predictions = lr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
new = np.asarray(y_test)
confusion_matrix(predictions, y_test)
```
